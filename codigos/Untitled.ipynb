```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def save_path(name):
    return f'../img/{name}.png'
```

```python
xk = lambda k: k
def dac( xk, h ):
    result = []
    for xi in xk:
        result = [ *result, *(xi * np.ones(100)) ]
    return result

```

```python
k = np.arange(5)
xk_e = xk(k)
dac_e = dac(xk_e, 1)
t = np.linspace( 0, 5, len(dac_e) )
```

```python
plt.plot( t, dac_e )
plt.grid()
plt.savefig( save_path('dac') )
```

```python
plt.stem(k, xk_e)
plt.grid()
plt.savefig( save_path('xk') )
```
